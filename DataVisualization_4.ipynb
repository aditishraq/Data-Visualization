### Scatter Plots
A way to visaulize how two numeric variables are realated in your data 
- Relationship between multiple variables
- Helps you find outliers

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

#### Find relationship between GDP per capita and Life expectency 

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
plt.scatter(data_2007.gdpPercap, data_2007.lifeExp,5 )
plt.title('GDP per capita and life expectency in 2007')
plt.xlabel('GDP per capita($)')
plt.ylabel('life exceptency')
plt.show()
```

```python
data_2007.gdpPercap.corr(data_2007.lifeExp)
#corr- correlation (between GPD per capita & life expectency)
```

```python
np.log10([10, 100, 1000])
#numpy function computes the log of given number with base 10
```

```python
plt.scatter(np.log10(data_2007.gdpPercap), data_2007.lifeExp)
plt.show()
```

```python
np.log10(data_2007.gdpPercap).corr(data_2007.lifeExp)
```

```python
data_2007.gdpPercap.corr(data_2007.lifeExp)
```

```python
years_sorted = sorted(set(data.year))
```

```python
for given_year in years_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(data_year.gdpPercap, data_year.lifeExp ,5)
    plt.title(given_year)
    plt.xlim(0, 60000)
    plt.ylim(25, 85)
    plt.xlabel('GDP Per Capita')
    plt.ylabel('Life Expectancy')
    #plt.show()
    plt.savefig(str(given_year),dpi = 200)
    plt.clf()
```

```python
data[data.gdpPercap > 60000] #outlier
```

```python
for given_year in years_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(np.log10(data_year.gdpPercap), data_year.lifeExp ,5)
    plt.title(given_year)
    plt.xlim(2, 5)
    plt.ylim(25, 85)
    plt.xlabel('GDP Per Capita (log scale)')
    plt.ylabel('Life Expectancy')
    plt.show()
    #lt.savefig('log' + str(given_year),dpi = 200)
    plt.clf()
```

### Examine relationship between GDP (NOT per capita) and life expectency in 2007
- Find GDP with data.gdpPercap * data.population
- Make a scatter plot
- Use both raw GDP and log og GDP

```python
data_2007 = data[data.year == 2007]
gdp_in_billions = (data_2007.gdpPercap * data_2007.population) / 10**9)
plt.scatter(gdp_in_billions, data_2007.lifeExp ,5)
plt.show()
```

```python
plt.scatter(np.log10(gdp_in_billions), data_2007.lifeExp ,5)
plt.show()
```

```python
plt.scatter(np.log10(data_2007.gdpPercap), data_2007.lifeExp, 5)
plt.show()
```

```python
print('Correlation between log10 of GDP and life expectency: ')
print(np.log10(gdp_in_billions).corr(data_2007.lifeExp))
print('Correlation between log10 of GDP per capita and life expectency: ')
print(np.log10(data_2007.gdpPercap).corr(data_2007.lifeExp))
```
