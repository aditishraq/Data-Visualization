### Is Kuwait really the rickhest country in the world (on a per-person basis)

```python
kuwait = data[data.country == 'Kuwait']
kuwait.head()
```

```python
plt.plot(kuwait.year, kuwait.gdpPercap)
plt.title('GDP per capita in kuwait')
plt.show()
```

```python
plt.subplot(311)
plt.title('GDP per capita')
plt.plot(kuwait.year, kuwait.gdpPercap)

plt.subplot(312)
plt.title('GDP in billions')
plt.plot(kuwait.year, kuwait.population * kuwait.gdpPercap / 10**9)

plt.subplot(313)
plt.title('Population in millions')
plt.plot(kuwait.year, kuwait.population / 10**6)

plt.tight_layout()
plt.show()
```

```python
plt.plot(kuwait.year, kuwait.population / kuwait.population.iloc[0] * 100)

kuwait_gdp = kuwait.population * kuwait.gdpPercap
plt.plot(kuwait.year, kuwait_gdp / kuwait_gdp.iloc[0] * 100)
plt.plot(kuwait.year, kuwait.gdpPercap / kuwait.gdpPercap.iloc[0] * 100)
plt.title('GDP and Population Growth in Kuwait (first year = 100)')
plt.legend(['Population', 'GDP','GDP Per Capita'])

plt.tight_layout()
plt.show()
```
